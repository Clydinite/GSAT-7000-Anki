```python
# insert a column before the first (1st) column

import pandas as pd

df = pd.read_csv('raw_gsat_data.tsv', sep='\t')

with open('data/vocabulary/level3.txt', 'r', encoding='utf-8') as f:
    words = list(line.strip() for line in f) 
    
df.insert(1, 'raw_string', '')

for index, row in df.iterrows():
    df.at[index, 'raw_string'] = words.pop(0) if words else ''
    
df.to_csv('raw_gsat_data_with_raw_string.tsv', sep='\t', index=False)
```
